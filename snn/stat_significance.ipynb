### Normality Check - diff AUC Values

```python
from scipy.stats import shapiro
import numpy as np

data_types = ["Morgan", "MACCS","RDKit", "MACCS + Morgan", "Morgan + RDKit"]

for task_idx, task in enumerate(tasks):
    print(task)
    for data_idx, data_type in enumerate(data_types):
        # Baselines
        print("."*4, data_type)
        for b_idx, baseline in enumerate(results_baselines):
            diff = np.array(results_CSNN[task_idx][data_idx]) - np.array(baseline[task_idx][data_idx])
            stat, p = shapiro(diff)
            reject = "Reject H0" if p < 0.05 else "Fail to reject H0"
            print("."*8, baseline_names[b_idx]," p=", p, "->", reject)

```

### Wilcoxon Test

```python
from scipy.stats import wilcoxon

# Headers
print("WILCOXON")

for task_idx, task in enumerate(tasks):
    print(task)
    print("Model", end='')
    for data_type in data_types:
        print(f" | {data_type}", end='')
    print()
    for b_idx, baseline in enumerate(results_baselines):
        if baseline_names[b_idx] == "RF":
            string = f"{baseline_names[b_idx]} "
        else: string = f"{baseline_names[b_idx]}"
        print(string, end='')

        for data_idx, data_type in enumerate(data_types):
            diff = np.array(results_CSNN[task_idx][data_idx]) - np.array(baseline[task_idx][data_idx])
            stat, p = wilcoxon(diff)
            print(f" & {p:.3e}", end='')
        print()
```
